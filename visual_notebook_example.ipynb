```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Hurricane Dorian Example

```python
df = pd.read_csv('master.csv')
df.head()

df2 = pd.read_csv('dorian_df.csv')

df2.drop(columns = ['zip', 'StateName'], inplace = True)
df2.head()
```

```python
df['2019-07'] = df2['2019-07']
df['2019-08'] = df2['2019-08']
df['2019-09'] = df2['2019-09']
df['2019-10'] = df2['2019-10']
df['dorian_affected'] = df2['dorian_affected']
df['%_change_after_dorian'] = df2['%_change_after_dorian']

df.head()
```

```python
dorian = df[df['dorian_affected'] == 1]
dorian.drop(columns =['sandy_affected', '%_change_after_sandy', 'harvey_affected'], inplace = True)
dorian.head()
```

```python
#mean %_change_after_dorian
dorian['%_change_after_dorian'].mean()
```

We note that the mean percent changed because of hurricane Dorian wasn't that substantial.

```python
dorian['%_change_after_dorian'].max()
```

```python
dorian[dorian['%_change_after_dorian'] == 0.2677469135802469]
```

The max positive percent change is around 27% in the city of Eustis. 

```python
dorian['%_change_after_dorian'].min()
```

```python
dorian[dorian['%_change_after_dorian'] == -0.2079318013343217]
```

The max negative percent change is around -21% in the city of Daytona beach

```python
plt.scatter(dorian['zip'], dorian['%_change_after_dorian']);
#plt.savefig('percent_change_scatter.png')
```

This scatterplot shows that there doesn't seem to be that much of an impact on property value.

```python
#I wanted to include the min/max percent change zipcodes
#row w/ max % change
plt.figure(figsize=(12,8))
plt.title('Mean_Median values for 5 zip codes affected by Dorian')
row = dorian.loc[6198]
plt.plot(row[3:15]);
#row w/ min % change
row2 = dorian.loc[6487]
plt.plot(row2[3:15]);
row3 = dorian.loc[3226]
plt.plot(row3[3:15]);
row4 = dorian.loc[1570]
plt.plot(row4[3:15]);
row5 = dorian.loc[2200]
plt.plot(row5[3:15]);
plt.legend( ['Eustis', 'Daytona Beach', 'Hollywood', 'River Park', 'Miami']);
plt.xticks(rotation = 45);
plt.tight_layout;
#plt.savefig('dorian_trends.png', bbox_inches="tight")
```

Couple things to note here: 
- the NA values in the dataset make it hard to interpret overall trends
- mostly random sample of zipcodes display very different trends
- each zipcode has a certain range, some bigger than others
- can't see the min/max percent change cuz this plots years, not months

# Narrowing down to the months near the occurrence of the hurricane

```python
df = pd.read_csv('Sale_median_Prices_ZipCode_2_minor_cleaning.csv')
df.drop(columns = ['Unnamed: 0'], inplace = True)
zips = pd.read_csv('Dorian_zip.csv')
```

```python
df['zip'] = df['zip'].astype(str)
df['zip'] = df['zip'].str[:5]

zips['0'] = zips['0'].astype(str)
```

```python
dorian = []
for i in df['zip']:
    if i in set(zips['0']):
        dorian.append(1)
    else:
        dorian.append(0)
```

```python
df['dorian_affected'] = dorian

df2 = df[df['dorian_affected']==1]
```

```python
df2.sample(5)
```

```python
plt.figure(figsize=(12,8))
plt.title('Monthly Median values for Dorian')
#max %change
row = df2.loc[6198]
plt.plot(row[140:146]);
row = df2.loc[3279]
plt.plot(row[140:146]);
row = df2.loc[2637]
plt.plot(row[140:146]);
#max negative %change
row = df2.loc[6487]
plt.plot(row[140:146]);
row = df2.loc[1137]
plt.plot(row[140:146]);
plt.legend( ['Eustis', 'Jacksonville', 'Melbourne', 'Daytona Beach', 'Orlando']);
plt.xticks(rotation = 45);
plt.tight_layout;
plt.savefig('dorian_month_trends.png', bbox_inches="tight")
```
